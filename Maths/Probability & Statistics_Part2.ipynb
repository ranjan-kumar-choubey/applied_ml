We have seen that , Given a random varibable which was log-normally distributed , We can easily converted it into Gaussian Distributed random

```python
#Example
from scipy import stats
import matplotlib.pyplot as plt

# We generate some random variates from a non-normal distribution(Gamma Distribution) and 
# make a probability plot for it, to show it is non-normal in the tails:

fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5, size=500) + 5
prob = stats.probplot(x, dist=stats.norm, plot=ax1) # this is Q-Q plot
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
print("\n")

# We now use boxcox to transform the data so it’s closest to normal:
ax2 = fig.add_subplot(212)
Y, _ = stats.boxcox(x) # it will convert in Gaussian Distribution
prob = stats.probplot(Y, dist=stats.norm, plot=ax2) # Q-Q plot
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()



# ref :https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html
```

- We Can See, using Q-Q Plot that Last Graph is Convert into Gaussian Distribution

#### Pearson Correlation coefficient (Rho) :
-  always lies bw [-1,+1],

- Rho = -1 ; X and Y have inverse relationship and All points lies on straight line

- Rho = (-1,0); X and Y have inverse relationship and points does not lies on straight line

- Rho = 0 ; X and Y have no relationship

- Rho = (0,1) ; X and Y have direct relationship and points does not lies on straight line

- Rho = +1 ; X and Y have direct relationship and All points lies on straight line




Most Of the time , It tells about linear relationship between two Random Variable

<ins>Limitation of Pearson Correlation coefficient

1. It doesn't tell about how fast a random variable increase with respect to other random variable. 
See below Example, In all cases , PCC = 1 and -1 ,while These curve has different rate of increasing & decreasing.



2. It is unable to explain  complex non-linear relationship bw random variable. It directly tells that PCC=0, while we can see that 
there has been some non-linear relationship between Random Varibale.



3. Pearson Correlation Coefficient is not good in non-linear relationship,  While Spearman's Rank correlation coefficient tells about  monotonic relationships (whether linear or not)

Ex : In case of Non-decreasing function 

    PCC = 0.8 ; It should be 1 ,but It gives value 0.8 while,
    Spearman's Rank Correlation coefficient = 1

- Advantage of Spearman's Rank Correlation Co-efficient over Pearson :
