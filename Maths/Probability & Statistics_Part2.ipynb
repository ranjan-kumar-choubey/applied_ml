- ### <ins> Box-Cox Transformation : 

We have seen that , Given a random varibable which was log-normally distributed , We can easily converted it into Gaussian Distributed random
varibale Y by taking natural log of all data points.

- In Machine Learning, Most of the time we try assume or to convert other random variable into Gaussian Random variable,Because We can derive many insights from it.




- Q. How Power Transform Works ?

    Ans : 
    
            Let 
                X : [x1, x2, x3 ... xn ] = A Pareto Distributed Random Varibale
                Y : [y1, y2, y3 ... yn]= Gaussian Distributed Random Varibale

            We want to convert this X into Y
 

Note : 
-        If lambda = 0 , It means that X is log-Normal Distributed, that's why we are just taking Natural Logrithm to Y in this condition(lambda=0) to convert into Gaussian Distribution.

- don't much worry about " lambda = box-cox(X) " It is out of scope

```python
#Example
from scipy import stats
import matplotlib.pyplot as plt

# We generate some random variates from a non-normal distribution(Gamma Distribution) and 
# make a probability plot for it, to show it is non-normal in the tails:

fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5, size=500) + 5
prob = stats.probplot(x, dist=stats.norm, plot=ax1) # this is Q-Q plot
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
print("\n")

# We now use boxcox to transform the data so its closest to normal:
ax2 = fig.add_subplot(212)
Y, _ = stats.boxcox(x) # it will convert in Gaussian Distribution
prob = stats.probplot(Y, dist=stats.norm, plot=ax2) # Q-Q plot
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()



# ref :https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html
```

- We Can See, using Q-Q Plot that Last Graph is Convert into Gaussian Distribution
